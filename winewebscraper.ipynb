Below is the data collection for some selected red wines from wine searcher


```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Create empty lists to store data
wine_names_list = []
ratings_list = []
regions_list = []
price_list = []

n=12 #n is number of wines on the page
# Loop through each URL in the "Reds WS" column
for url in urls['Reds WS']:
  try:
    response = requests.get(url)
    # check the status code to see if the request was successful
    if response.status_code == 200:
        html_content = response.content
        # Use BeautifulSoup to parse the HTML content
        soup = BeautifulSoup(html_content, "html.parser")
        #print(soup)
        wine_names = [wine.text.strip() for wine in soup.find_all("div", {"class": "product-name"})[:n]]
        ratings =  [wine.text.strip() for wine in soup.find_all("span", {"class": "badge badge-pill badge-rating"})[:n]]
        regions = [wine.text.strip() for wine in soup.find_all("td", {"class": "align-middle text-lg-center"})[:n]]
        price = [wine.text.strip() for wine in soup.find_all("div", {"class": "price text-right text-lg-left"})[:n]]
        wine_names_list += wine_names
        ratings_list += ratings
        regions_list += regions

        price_list += price
    else:
        pass
  except requests.exceptions.RequestException as e:
    pass


# Create a dictionary with the extracted data
data_dict = {"wine_names": wine_names_list,
             "ratings": ratings_list,
             "regions": regions_list,
             "price": price_list}
def strip_text(column):
    return column.apply(lambda x: x.split("\n\n")[0].strip())
def drop_last_n(column, n):
    return column.apply(lambda x: x[:-n])
# Convert the dictionary into a pandas dataframe
print(data_dict)
df = pd.DataFrame(data_dict)
df['wine_names'] = strip_text(df['wine_names'])
df['regions'] = strip_text(df['regions'])
df['price'] = drop_last_n(df['price'],13)
df.to_csv('assortedRedsWineSearcher.csv')
```
