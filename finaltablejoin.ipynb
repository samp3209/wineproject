```python
import pandas as pd
```

```python
df = pd.read_csv('winedf - winedf (1).csv')
newcols = pd.read_csv('Wine Vintage Years - newcols.csv')
```

```python
df = pd.merge(df, newcols[['Vintage', 'Drinkability', 'Desirability']],
              how='left', left_on=['year'], right_on=['Vintage'])
```

```python
df.drop('Vintage', axis=1, inplace=True)
```
