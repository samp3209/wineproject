```python
import pandas as pd
```

```python
df = pd.read_csv('wine - df.csv')
cali_df = pd.read_csv("cali.csv")
italy_df = pd.read_csv("italy.csv")
```

```python
# assuming your main dataframe is called 'df' and the california dataframe is called 'cali_df'
# merging the dataframes based on 'year' column
df = pd.merge(df, cali_df[['Vintage', 'Temperature', 'cali', 'cali points', 'season rating']], 
              how='left', left_on=['year'], right_on=['Vintage'])

# filling the NaN values with 0 in 'cali' and 'cali points' columns for wines not from California
df.loc[df['country'] != 'cali', ['cali', 'cali points']] = 0
df.loc[df['country'] != 'cali', ['cali', 'Temperature']] = 0

# renaming the columns
df = df.rename(columns={'Temperature': 'cali temperature', 
                        'cali': 'cali region rating', 
                        'cali points': 'cali wine points', 
                        'season rating': 'total season rating'})

# dropping the 'Vintage' column
df.drop('Vintage', axis=1, inplace=True)
```

```python
df = pd.merge(df, italy_df[['Vintage', 'Temperature', 'italy', 'italy points']],
              how='left', left_on=['year'], right_on=['Vintage'])

df.loc[df['country'] != 'italy', ['italy', 'italy points']] = 0
df.loc[df['country'] != 'italy', ['italy', 'Temperature']] = 0 

df = df.rename(columns={'Temperature': 'italy temperature',
                        'italy': 'italy region rating'})
df.drop('Vintage', axis=1, inplace=True)
```

```python
germany_df = pd.read_csv('germany.csv')
df = pd.merge(df, germany_df[['Vintage', 'Temperature', 'germany', 'germany points']],
              how='left', left_on=['year'], right_on=['Vintage'])

df.loc[df['country'] != 'germany', ['germany', 'germany points']] = 0
df.loc[df['country'] != 'germany', ['germany', 'Temperature']] = 0 

df = df.rename(columns={'Temperature': 'germany temperature',
                        'germany': 'germany region rating'})
df.drop('Vintage', axis=1, inplace=True)
```
