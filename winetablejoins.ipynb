```python
import pandas as pd
```

```python
df = pd.read_csv('wine - df.csv')
cali_df = pd.read_csv("cali.csv")
```

```python
# assuming your main dataframe is called 'df' and the california dataframe is called 'cali_df'
# merging the dataframes based on 'year' column
df = pd.merge(df, cali_df[['Vintage', 'Temperature', 'cali', 'cali points', 'season rating']], 
              how='left', left_on=['year'], right_on=['Vintage'])

# filling the NaN values with 0 in 'cali' and 'cali points' columns for wines not from California
df.loc[df['country'] != 'california', ['cali', 'cali points']] = 0

# renaming the columns
df = df.rename(columns={'Temperature': 'cali temperature', 
                        'cali': 'cali region rating', 
                        'cali points': 'cali wine points'})

# dropping the 'Vintage' column
df.drop('Vintage', axis=1, inplace=True)
```
